```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self,x):
        self.val = x
        self.left = None
        self.right = None
        """
        :type val: int
        :type left: TreeNode or None
        :type right: TreeNode or None
        """
    def Printtree(self):
        
        if self.left:
            self.left.Printtree()
            
        print(self.val),
        
        if self.right:
            self.right.Printtree()
            
class Solution(object):
    
    def insert(self, root, val):
        """
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode(inserted node)
        """
        data = TreeNode(val)
        
        if root.val:    #root.val å­˜åœ¨ï¼Œå¾€ä¸‹èµ°
            if val <= root.val:
                
                if root.left is None:
                    root.left = data
                    return root.left            # å¦‚æœå·¦é‚Šçš„ç¯€é»æ˜¯ç©ºçš„ï¼Œé‚£éº¼æŠŠæ–°çš„å€¼æ”¾é€²å»
                else:
                    return self.insert(root.left,val)   # å¦‚æœå·¦é‚Šç›®å‰æ˜¯æœ‰å€¼å­˜åœ¨çš„è©±ï¼Œå†æ¬¡å‘¼å«insertï¼Œç¹¼çºŒå¾€å·¦é‚Šèµ°
                
            elif val > root.val:
               
                if root.right is None:    
                    root.right = data
                    return root.right
                else:
                    return self.insert(root.right,val)   # åŒå³é‚Šå¯«æ³•
                
    def delete(self, root, target):
        """
        :type root: TreeNode
        :type target: int
        :rtype: TreeNode(the root of new completed binary search tree) (cannot search())
        """
        
        node = self.search(root,target)   #å…ˆæ‰¾çœ‹çœ‹è¦åˆªé™¤çš„æ•¸æ˜¯å¦å­˜åœ¨
        parent = root    #éœ€è¦ä¸€å€‹æœƒç§»å‹•çš„parentï¼Œæœ‰é»åƒå“¨å…µ
        
        if root == None:
            return root      #å¦‚æœæ²’æœ‰rootå›å‚³
        if node:
            if target < root.val:
                root.left = self.delete(root.left,target)
            elif target > root.val:
                root.right = self.delete(root.right,target)
                
                #æ‰¾targetï¼Œåœ¨rootå·¦é‚Šæˆ–å³é‚Šï¼Œæ‰¾åˆ°äº†çš„è©±å†æ¬¡å‘¼å«delete functionï¼Œ

            else:

                if root.left is None and root.right is None:
                    
                    root = None  
                
                # 0å€‹å°å­©çš„æƒ…æ³ï¼Œç•¶å·¦é‚Šå’Œå³é‚Šæ˜¯ç©ºç¯€é»æ™‚ï¼Œç•¶ä¸‹çš„rootå°‡æœƒæˆç‚ºç©ºå€¼

                elif root.left and (root.right is None):                    
                    root = root.left
                    self.delete(root,target)

                elif (root.left is None) and root.right:
                    root = root.right
                    self.delete(root,target)
                    
                # 1å€‹å°å­©çš„æƒ…æ³ï¼Œè‹¥å·¦é‚Šæˆ–å³é‚Šåªæœ‰ä¸€å€‹ç¯€é»æ˜¯å­˜åœ¨çš„è©±ï¼Œæœƒå°‡æ–°çš„rootè³œäºˆçµ¦é‚£ä½å°å­©ï¼Œä¸¦çœ‹çœ‹ä»–æœ‰æ²’æœ‰æ›´å¤šå°å­©ï¼Œè®“éˆæ¥ä¸æœƒæ–·æ‰

                elif root.left and root.right:

                    parent = root
                    successor = root.right

                    while successor.left:
                        parent = successor
                        successor = successor.left

                    root.val = successor.val

                    if root == successor:
                        parent.left = successor.right
                    else:
                        parent.right = successor.right
                
                # 2å€‹å°å­©çš„æƒ…æ³æœ‰å…©ç¨®æ–¹æ³•ï¼Œä¸€ç¨®æ˜¯çœ‹å·¦é‚Šï¼ˆleft subtree)ï¼Œæ‰¾å‡ºæœ€å¤§å€¼ï¼ˆpredecessor)ï¼›ä¸€ç¨®æ˜¯çœ‹å³é‚Šï¼ˆright subtree)ï¼Œæ‰¾å‡ºæœ€å°å€¼ï¼ˆsuccessor)
                # æœ€å¾Œç”¨parentå’Œsuccessorå»èµ°å‹•ï¼Œå°‡right subtreeä¸­æœ€å·¦é‚Šçš„æœ€å°å€¼=successorï¼Œè·Ÿrootäº¤æ›ï¼Œé€™æ™‚å€™parentæœƒæ˜¯successorçš„çˆ¶ç¯€é»ï¼Œå› ç‚ºä¸èƒ½ç¢ºå®šsuccessoræœ‰æ²’æœ‰å°å­©ï¼Œä½†æ˜¯åªæœƒæœ‰ä¸€ç¨®æƒ…æ³ï¼š
                # å³é‚Šæœ‰å°å­©ï¼ˆå› ç‚ºå¦‚æœå·¦é‚Šæœ‰å°å­©ï¼Œé‚£éº¼successorå°±ä¸æˆç«‹ï¼‰ï¼Œç•¶successoræˆç‚ºrootæ™‚ï¼Œä»–çš„å°å­©æœƒæ¥ä¸Šparent<æœ‰é»åƒæ˜¯çˆ¸çˆ¸å»ä¸Šç­ï¼Œå­©å­çµ¦çˆºçˆºå¸¶>
                

            return root

    def search(self, root, target):
        """
        :type root: TreeNode
        :type target: int
        :rtype: TreeNode(searched node)
        """
        if target < root.val:
            if root.left == None:
                return None 
            return self.search(root.left,target)
        
        elif target > root.val:
            if root.right == None:
                return None
            return self.search(root.right,target)
        
        else:
            return root
        
        # æŸ¥è©¢targetæ˜¯å¦æœ‰åœ¨é€™å€‹æ£µæ¨¹è£¡ï¼Œæ²’æœ‰å›å‚³Noneï¼Œæœ‰çš„è©±å›å‚³root
        
    def modify(self, root, target, new_val):
        """
        :type root: TreeNode
        :type target: int
        :type new_val: int
        :rtype:TreeNode(the root of new completed binary search tree) (cannot search())
        """
        node = self.search(root,target)
        i = root
        
        while node != None:
    
            if target < root.val:
                root.left = self.delete(root.left,target)
            elif target > root.val:
                root.right = self.delete(root.right,target)

            else:

                if root.left is None and root.right is None:
                    
                    root = None

                elif root.left and (root.right is None):                    
                    root = root.left
                    self.delete(root,target)

                elif (root.left is None) and root.right:
                    root = root.right
                    self.delete(root,target)

                elif root.left and root.right:

                    parent = root
                    successor = root.right

                    while successor.left:
                        parent = successor
                        successor = successor.left

                    root.val = successor.val

                    if root == successor:
                        parent.left = successor.right
                    else:
                        parent.right = successor.right
                
            self.insert(i,new_val)
            return i
        
        # modifyæˆ‘ä¸å¤ªæ‡‚é€™å€‹è¦æ€éº¼åšåˆ°ä¿®æ”¹å¾Œï¼Œæ¨¹é«˜æœƒå°æ–¼ç­‰æ–¼ä¹‹å‰çš„ï¼Œæ‰€ä»¥æˆ‘ç”¨çš„æ–¹æ³•æ˜¯å…ˆå‘¼å«deleteï¼Œå°‡targetåˆªé™¤ï¼Œå†insertå°‡new_valæ”¾å…¥ã€‚
        
```

```python
class MyBST(object):
    
    def __init__(self,val):
        
        self.rightkid = None
        self.leftkid = None
        self.val = val
        
    def PrintTree(self):
            
        if self.leftkid:
            self.leftkid.PrintTree()
        
        print(self.val) 
        
        if self.rightkid:
            self.rightkid.PrintTree()
        
    def insert(self,root,val):
        
        if root.val:
            
            if val <= root.val:
                
                if root.leftkid is None:
                    root.leftkid = MyBST(val)
                else:
                    root.leftkid.insert(root.leftkid,val)   
                
            elif val > self.val:
               
                if root.rightkid is None:
                    root.rightkid = MyBST(val)
                else:
                    root.insert(root.rightkid,val) 
        
    def search(self, root, target):
        
        if target < root.val:

            if root.leftkid is None:
                return None
            return self.search(root.leftkid,target)

        elif target > root.val:

            if root.rightkid is None:
                return None
            return self.search(root.rightkid,target)

        else:
            return root

    def delete(self, root, target):
            
        if root.val > target:
            root.leftkid = self.delete(root.leftkid, target)

        elif root.val < target:
            root.rightkid = self.delete(root.rightkid, target)

        else:
            
            if root.rightkid == None and root.leftkid == None:
                
                root = None

            elif not root.rightkid:
                
                root = root.leftkid 
                self.delete(root,target)

            elif not root.leftkid:
                
                root = root.rightkid 
                self.delete(root,target)

            
            elif root.rightkid and root.leftkid:
                
                global successor
                parent = root
                successor = root.rightkid
            
                while successor.leftkid:
                    parent = successor
                    successor = successor.leftkid
                    

                root.val = successor.val

                if successor == root:
                    parent.leftkid = root.rightkid
                else:
                    parent.rightkid = root.rightkid
            
        return root
```

```python
root = MyBST(6)
Node1 = MyBST(10)
Node2 = MyBST(5)
Node3 = MyBST(-4)
Node4 = MyBST(7) 
root.rightkid = Node1
root.leftkid = Node2
Node2.leftkid = Node3
Node1.leftkid = Node4
root.insert(root,5)
root.insert(root,100)
root.insert(root,-5)
root.PrintTree()
```
