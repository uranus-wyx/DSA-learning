```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self,x):
        self.val = x
        self.left = None
        self.right = None
        """
        :type val: int
        :type left: TreeNode or None
        :type right: TreeNode or None
        """
    def Printtree(self):
        
        if self.left:
            self.left.Printtree()
            
        print(self.val),
        
        if self.right:
            self.right.Printtree()
            
class Solution(object):
    
    def insert(self, root, val):
        """
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode(inserted node)
        """
        data = TreeNode(val)
        
        if root.val:    #root.val å­˜åœ¨ï¼Œå¾€ä¸‹èµ°
            if val <= root.val:
                
                if root.left is None:
                    root.left = data
                    return root.left            # å¦‚æœå·¦é‚Šçš„ç¯€é»æ˜¯ç©ºçš„ï¼Œé‚£éº¼æŠŠæ–°çš„å€¼æ”¾é€²å»
                else:
                    return self.insert(root.left,val)   # å¦‚æœå·¦é‚Šç›®å‰æ˜¯æœ‰å€¼å­˜åœ¨çš„è©±ï¼Œå†æ¬¡å‘¼å«insertï¼Œç¹¼çºŒå¾€å·¦é‚Šèµ°
                
            elif val > root.val:
               
                if root.right is None:    
                    root.right = data
                    return root.right
                else:
                    return self.insert(root.right,val)   # åŒå³é‚Šå¯«æ³•
                
    def delete(self, root, target):
        """
        :type root: TreeNode
        :type target: int
        :rtype: TreeNode(the root of new completed binary search tree) (cannot search())
        """
        
        node = self.search(root,target)   #å…ˆæ‰¾çœ‹çœ‹è¦åˆªé™¤çš„æ•¸æ˜¯å¦å­˜åœ¨
        parent = root    #éœ€è¦ä¸€å€‹æœƒç§»å‹•çš„parentï¼Œæœ‰é»åƒå“¨å…µ
        
        if root == None:
            return root      #å¦‚æœæ²’æœ‰rootå›å‚³
        if node:
            if target < root.val:
                root.left = self.delete(root.left,target)
            elif target > root.val:
                root.right = self.delete(root.right,target)
                
                #æ‰¾targetï¼Œåœ¨rootå·¦é‚Šæˆ–å³é‚Šï¼Œæ‰¾åˆ°äº†çš„è©±å†æ¬¡å‘¼å«delete functionï¼Œ

            else:

                if root.left is None and root.right is None:
                    
                    root = None  
                
                # 0å€‹å°å­©çš„æƒ…æ³ï¼Œç•¶å·¦é‚Šå’Œå³é‚Šæ˜¯ç©ºç¯€é»æ™‚ï¼Œç•¶ä¸‹çš„rootå°‡æœƒæˆç‚ºç©ºå€¼

                elif root.left and (root.right is None):                    
                    root = root.left
                    self.delete(root,target)

                elif (root.left is None) and root.right:
                    root = root.right
                    self.delete(root,target)
                    
                # 1å€‹å°å­©çš„æƒ…æ³ï¼Œè‹¥å·¦é‚Šæˆ–å³é‚Šåªæœ‰ä¸€å€‹ç¯€é»æ˜¯å­˜åœ¨çš„è©±ï¼Œæœƒå°‡æ–°çš„rootè³œäºˆçµ¦é‚£ä½å°å­©ï¼Œä¸¦çœ‹çœ‹ä»–æœ‰æ²’æœ‰æ›´å¤šå°å­©ï¼Œè®“éˆæ¥ä¸æœƒæ–·æ‰

                elif root.left and root.right:

                    parent = root
                    successor = root.right

                    while successor.left:
                        parent = successor
                        successor = successor.left

                    root.val = successor.val

                    if root == successor:
                        parent.left = successor.right
                    else:
                        parent.right = successor.right
                
                # 2å€‹å°å­©çš„æƒ…æ³æœ‰å…©ç¨®æ–¹æ³•ï¼Œä¸€ç¨®æ˜¯çœ‹å·¦é‚Šï¼ˆleft subtree)ï¼Œæ‰¾å‡ºæœ€å¤§å€¼ï¼ˆpredecessor)ï¼›ä¸€ç¨®æ˜¯çœ‹å³é‚Šï¼ˆright subtree)ï¼Œæ‰¾å‡ºæœ€å°å€¼ï¼ˆsuccessor)
                # æœ€å¾Œç”¨parentå’Œsuccessorå»èµ°å‹•ï¼Œå°‡right subtreeä¸­æœ€å·¦é‚Šçš„æœ€å°å€¼=successorï¼Œè·Ÿrootäº¤æ›ï¼Œé€™æ™‚å€™parentæœƒæ˜¯successorçš„çˆ¶ç¯€é»ï¼Œå› ç‚ºä¸èƒ½ç¢ºå®šsuccessoræœ‰æ²’æœ‰å°å­©ï¼Œä½†æ˜¯åªæœƒæœ‰ä¸€ç¨®æƒ…æ³ï¼š
                # å³é‚Šæœ‰å°å­©ï¼ˆå› ç‚ºå¦‚æœå·¦é‚Šæœ‰å°å­©ï¼Œé‚£éº¼successorå°±ä¸æˆç«‹ï¼‰ï¼Œç•¶successoræˆç‚ºrootæ™‚ï¼Œä»–çš„å°å­©æœƒæ¥ä¸Šparent<æœ‰é»åƒæ˜¯çˆ¸çˆ¸å»ä¸Šç­ï¼Œå­©å­çµ¦çˆºçˆºå¸¶>
                

            return root

    def search(self, root, target):
        """
        :type root: TreeNode
        :type target: int
        :rtype: TreeNode(searched node)
        """
        if target < root.val:
            if root.left == None:
                return None 
            return self.search(root.left,target)
        
        elif target > root.val:
            if root.right == None:
                return None
            return self.search(root.right,target)
        
        else:
            return root
        
        # æŸ¥è©¢targetæ˜¯å¦æœ‰åœ¨é€™å€‹æ£µæ¨¹è£¡ï¼Œæ²’æœ‰å›å‚³Noneï¼Œæœ‰çš„è©±å›å‚³root    
    
    def modify(self, root, target, new_val):
        """
        :type root: TreeNode
        :type target: int
        :type new_val: int
        :rtype:TreeNode(the root of new completed binary search tree) (cannot search())
        """
        old = self.Tree_array(root)
        new = []
        for i in old:
            if i == target:
                i = new_val
                new.append(i)
            else:
                new.append(i)
        
        new_array = self.heap_sort(new)
        new_root = len(new_array)//2
        new_data = TreeNode(new_array[new_root])
        
        for j in range(len(new_array)):
            if j != new_root:
                self.insert(new_data, new_array[j])

        return new_data
    
      #å…ˆå°‡æœ¬ä¾†çš„æ¨¹è®Šæˆä¸€å€‹listï¼Œç„¶å¾Œå¦‚æœè£¡é¢ç¬¬iå€‹ä½ç½®çš„å€¼=target(ç›®æ¨™åˆªé™¤ï¼‰ï¼Œä»¥new_valå–ä»£ï¼Œå…¶ä»–å‰‡ä¸€ä¸€appendè‡³new[]è£¡
      #new[]éœ€è¦å…ˆheapsortï¼Œç”±å°æ’åˆ—åˆ°å¤§ï¼Œå†å–å‡ºä¸­é–“å€¼ï¼Œç¢ºä¿æ–°çš„æ¨¹çš„æ¨¹é«˜æœƒå°æ–¼ç­‰æ–¼åŸæœ¬çš„æ¨¹
      #new_dataæœƒæˆç‚ºæ–°çš„rootï¼Œç„¶å¾Œå†ç”¨ä¸€å€‹forè¿´åœˆï¼ŒæŠŠnew_arrayä¸­é™¤new_rootçš„å€¼æ”¾å…¥binary treeè£¡ï¼ŒçµæŸã€‚
        
    
    def Tree_array(self, root):
        
        vals = []

        def dfs(node):
            if node:
                vals.append(node.val)
                dfs(node.left)
                dfs(node.right)

        dfs(root)
        return vals
    #å¼•ç”¨univaltreeçš„æ–¹æ³•ï¼Œå°‡nodeæ”¾é€²vals[]è£¡ï¼Œæª”æ¡ˆåœ¨anacondaï¼Œé‚„æœªä¸Šå‚³
    
    def heap_sort(self,nums):
        n = len (nums)    # n æ˜¯ numsæ•¸åˆ—é•·åº¦
        parent = (n//2)-1  #æœ€å¾Œä¸€å€‹çˆ¶ç¯€é»

        while parent >=0:  #ç•¶çˆ¶ç¯€é»>=0æ™‚ï¼Œç¹¼çºŒåŸ·è¡Œ
            self.heapify(nums,parent,n)  #å‘¼å«Solutionä¸­çš„heapifyå‡½æ•¸ï¼Œå°‡æ‰€æœ‰çˆ¶ç¯€é»>å­ç¯€é»
            parent-=1  #parentæœƒæ¸›å°‘ï¼Œå› ç‚ºè¦å¾æœ€ä¸‹é¢çš„ç¯€é»é–‹å§‹ç¯©é¸ï¼Œä¸€ä¸€å¾€ä¸Š

        end = len(nums)-1  #è¨­endæ˜¯æ•¸åˆ—é•·åº¦æ¸›1ï¼Œå› ç‚ºpythonçš„æ•¸åˆ—æ˜¯å¾0é–‹å§‹ï¼Œæ‰€ä»¥çµæŸçš„ä½ç½®æ˜¯len(nums)-1

        while (end > 0):  #ç•¶endå¤§æ–¼0æ™‚ï¼Œä¸‹ä¸€æ­¥
            nums[end] , nums[0] = nums[0] , nums[end]  #æ­¤æ™‚å› ç‚ºä¸Šæ–¹å·²ç¶“å‘¼å«äº†heapifyå‡½æ•¸ï¼Œæ‰€ä»¥index(0)æœƒæ˜¯æœ€å¤§çš„å€¼ï¼Œè·Ÿæœ€å¾Œä¸€å€‹å…ƒç´ çš„å€¼äº¤æ›
            self.heapify(nums, 0, end) #æ­¤æ­¥é©Ÿæ˜¯å› ç‚ºä¸Šä¸€æ­¥çš„äº¤æ›ä½¿å¾—index(0)ä¸æ˜¯æ•¸åˆ—ä¸­æœ€å¤§çš„å€¼ï¼Œæ‰€ä»¥å†æ¬¡å‘¼å«heapifyå‡½æ•¸
            end -= 1  #æ¯ä¸€æ¬¡çš„è¿´åœˆä¸­æœƒæŠŠæœ€å¤§çš„å€¼å¾€æœ€å¾Œä¸€å€‹å…ƒç´ æ”¾ï¼Œæ‰€ä»¥ä¸‹ä¸€æ¬¡è·‘è¿´åœˆæ™‚ï¼Œä¸å¯åŒ…å«æœ€å¾Œé¢å·²ç¶“äº¤æ›å®Œæˆçš„å…ƒç´ ï¼Œå› ç‚ºå®ƒå€‘å·²æ’åˆ—å¥½
                
        return nums #å›å‚³nums

    def heapify(self,nums, i, n):  #é€™å€‹å‡½æ•¸æ˜¯ç‚ºäº†è®“heapsortå‘¼å«è€Œè¨­çš„

        daddy = i
        leftkid = 2*i+1
        rightkid = 2*i+2  #çˆ¸çˆ¸ã€å·¦é‚Šå°å­©ã€å³é‚Šå°å­©ï¼Œçˆ¸çˆ¸>å·¦é‚Šå°å­©ã€å³é‚Šå°å­©

        if leftkid < n and nums[daddy] < nums[leftkid]:
            daddy = leftkid
        if rightkid < n and nums[daddy] < nums[rightkid]:
            daddy = rightkid
        if daddy != i:
            nums [i] ,nums[daddy] = nums[daddy] , nums[i]
            self.heapify(nums,daddy,n)
            
    #å¼•ç”¨è‡ªå·±çš„heap_sortæ–¹æ³•
```

```python
class MyBST(object):
    
    def __init__(self,val):
        
        self.rightkid = None
        self.leftkid = None
        self.val = val
        
    def PrintTree(self):
            
        if self.leftkid:
            self.leftkid.PrintTree()
        
        print(self.val) 
        
        if self.rightkid:
            self.rightkid.PrintTree()
        
    def insert(self,root,val):
        
        if root.val:
            
            if val <= root.val:
                
                if root.leftkid is None:
                    root.leftkid = MyBST(val)
                else:
                    root.leftkid.insert(root.leftkid,val)   
                
            elif val > self.val:
               
                if root.rightkid is None:
                    root.rightkid = MyBST(val)
                else:
                    root.insert(root.rightkid,val) 
        
    def search(self, root, target):
        
        if target < root.val:

            if root.leftkid is None:
                return None
            return self.search(root.leftkid,target)

        elif target > root.val:

            if root.rightkid is None:
                return None
            return self.search(root.rightkid,target)

        else:
            return root

    def delete(self, root, target):
            
        if root.val > target:
            root.leftkid = self.delete(root.leftkid, target)

        elif root.val < target:
            root.rightkid = self.delete(root.rightkid, target)

        else:
            
            if root.rightkid == None and root.leftkid == None:
                
                root = None

            elif not root.rightkid:
                
                root = root.leftkid 
                self.delete(root,target)

            elif not root.leftkid:
                
                root = root.rightkid 
                self.delete(root,target)

            
            elif root.rightkid and root.leftkid:
                
                global successor
                parent = root
                successor = root.rightkid
            
                while successor.leftkid:
                    parent = successor
                    successor = successor.leftkid
                    

                root.val = successor.val

                if successor == root:
                    parent.leftkid = root.rightkid
                else:
                    parent.rightkid = root.rightkid
            
        return root
```
