> Step2: è¨­è®Šæ•¸i, j, k çš†ç‚º0ï¼Œæ¥è‘—ä½¿ç”¨whileè¿´åœˆï¼Œ`while(æ¢ä»¶ï¼‰`ï¼Œå†ä¾†æœƒæ¯”è¼ƒ left[i] & right [j]ï¼Œå°çš„å…ƒç´ æœƒè¢«æ”¾é€²ä½ç½®ç‚ºkçš„æ–°arrayè£¡é¢ã€‚

> Step3: ä½†æ˜¯ç‚ºäº†é˜²æ­¢ç•¶å…¶ä¸­ä¸€é‚Šçš„é™£åˆ—è·‘å®Œè¿´åœˆï¼Œæ‰€ä»¥æ¢ä»¶è¦è¨­ä¸‰ç¨®ï¼š
>> ç¬¬ä¸€æ˜¯`i < len(left) and j < len(right)`  ï¼šå…©è€…éƒ½å°æ–¼è‡ªå·±çš„é•·åº¦æ™‚  
>> ç¬¬äºŒæ˜¯`i < len(left)`ï¼šå‡è¨­jå·²è·‘å®Œright

```python
def mergesort(array):
    if len(array)<=1:
        return array
    
    if len(array)>1:
        mid = len(array)//2
        left = array[:mid]
        right = array[mid:]
        
        mergesort(left)
        mergesort(right)

        i,j,k=0,0,0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                array[k] = left[i]
                i +=1
            else:
                array[k] = right[j]
                j+=1
            k+=1

        while i < len(left):
            array[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            array[k] = right[j]
            j+= 1
            k+= 1
            
        return array
```

```python
array = [3,1,5,8,2,7,9,4,6]
mergesort(array)
```
