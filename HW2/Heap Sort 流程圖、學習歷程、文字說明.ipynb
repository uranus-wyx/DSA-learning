```python
class Solution(object):
        
        def heap_sort(self,nums):
            n = len (nums)    # n æ˜¯ numsæ•¸åˆ—é•·åº¦
            parent = (n//2)-1  #æœ€å¾Œä¸€å€‹çˆ¶ç¯€é»

            while parent >=0:  #ç•¶çˆ¶ç¯€é»>=0æ™‚ï¼Œç¹¼çºŒåŸ·è¡Œ
                self.heapify(nums,parent,n)  #å‘¼å«Solutionä¸­çš„heapifyå‡½æ•¸ï¼Œå°‡æ‰€æœ‰çˆ¶ç¯€é»>å­ç¯€é»
                parent-=1  #parentæœƒæ¸›å°‘ï¼Œå› ç‚ºè¦å¾æœ€ä¸‹é¢çš„ç¯€é»é–‹å§‹ç¯©é¸ï¼Œä¸€ä¸€å¾€ä¸Š

            end = len(nums)-1  #è¨­endæ˜¯æ•¸åˆ—é•·åº¦æ¸›1ï¼Œå› ç‚ºpythonçš„æ•¸åˆ—æ˜¯å¾0é–‹å§‹ï¼Œæ‰€ä»¥çµæŸçš„ä½ç½®æ˜¯len(nums)-1

            while (end > 0):  #ç•¶endå¤§æ–¼0æ™‚ï¼Œä¸‹ä¸€æ­¥
                nums[end] , nums[0] = nums[0] , nums[end]  #æ­¤æ™‚å› ç‚ºä¸Šæ–¹å·²ç¶“å‘¼å«äº†heapifyå‡½æ•¸ï¼Œæ‰€ä»¥index(0)æœƒæ˜¯æœ€å¤§çš„å€¼ï¼Œè·Ÿæœ€å¾Œä¸€å€‹å…ƒç´ çš„å€¼äº¤æ›
                self.heapify(nums, 0, end) #æ­¤æ­¥é©Ÿæ˜¯å› ç‚ºä¸Šä¸€æ­¥çš„äº¤æ›ä½¿å¾—index(0)ä¸æ˜¯æ•¸åˆ—ä¸­æœ€å¤§çš„å€¼ï¼Œæ‰€ä»¥å†æ¬¡å‘¼å«heapifyå‡½æ•¸
                end -= 1  #æ¯ä¸€æ¬¡çš„è¿´åœˆä¸­æœƒæŠŠæœ€å¤§çš„å€¼å¾€æœ€å¾Œä¸€å€‹å…ƒç´ æ”¾ï¼Œæ‰€ä»¥ä¸‹ä¸€æ¬¡è·‘è¿´åœˆæ™‚ï¼Œä¸å¯åŒ…å«æœ€å¾Œé¢å·²ç¶“äº¤æ›å®Œæˆçš„å…ƒç´ ï¼Œå› ç‚ºå®ƒå€‘å·²æ’åˆ—å¥½
                
            return nums #å›å‚³nums

        def heapify(self,nums, i, n):  #é€™å€‹å‡½æ•¸æ˜¯ç‚ºäº†è®“heapsortå‘¼å«è€Œè¨­çš„

            daddy = i
            leftkid = 2*i+1
            rightkid = 2*i+2  #çˆ¸çˆ¸ã€å·¦é‚Šå°å­©ã€å³é‚Šå°å­©ï¼Œçˆ¸çˆ¸>å·¦é‚Šå°å­©ã€å³é‚Šå°å­©

            if leftkid < n and nums[daddy] < nums[leftkid]:
                daddy = leftkid
            if rightkid < n and nums[daddy] < nums[rightkid]:
                daddy = rightkid
            if daddy != i:
                nums [i] ,nums[daddy] = nums[daddy] , nums[i]
                self.heapify(nums,daddy,n)
            #ä¸Šé¢ä¸‰è¡Œifå°±æ˜¯çˆ¸çˆ¸å’Œå…’å­å€‘çš„æ•¸å€¼æ¯”è¼ƒï¼Œä¸‰è€…æœ€å¤§çš„indexä½ç½®å°‡æˆç‚ºçˆ¸çˆ¸çš„å»è™•ï¼Œå‰ææ˜¯å­©å­å€‘éƒ½ä¸å¾—å¤§æ–¼æ•¸åˆ—çš„é•·åº¦ï¼Œå› ç‚ºæœ‰çš„å­©å­å€‘ä¸¦æ²’æœ‰é–‹ææ•£è‘‰
            # ä½†ç•¶çˆ¸çˆ¸ç™¼ç¾æ‰‹è£¡çš„æ•¸å€¼ä¸¦ä¸æ˜¯åŸæœ¬è‡ªå·±çš„ï¼Œäº¤æ›æ•¸å€¼ï¼Œä½†æ˜¯é€™æ¨£ä¸å¤ ï¼Œå› ç‚ºä¸çŸ¥é“ä¸‹é¢çš„å­©å­å€‘æœƒä¸æœƒæœ‰å¯ä»¥æŠŠè‡ªå·±æ‰“è¶´çš„ï¼Œæ‰€ä»¥å†æ¬¡å‘¼å«å‡½æ•¸ï¼Œå¾€ä¸‹æ¯”è¼ƒ
```

```python
def maxheapify(array, n, i):
    root = i
    left = 2*i+1
    right = 2*i+2
    
    if left < n and array[root] < array[left]:
        root = left
        
    if right < n and array[root] <array[right]:
        root = right
        
    if root != i:
        array[i],array[root] = array[root],array[i]
        
        maxheapify(array, n, root)

def heapsort(array):
    n = len(array)
    
    for i in range(n, -1 , -1):
        maxheapify(array, n, i)
        
    for i in range (n-1, 0, -1):
        array[i] , array[0] = array[0] , array[i]
        maxheapify(array, i, 0)
```

```python
array = [3,5,6,4,1,7,2]
heapsort(array)
print (array)
```
