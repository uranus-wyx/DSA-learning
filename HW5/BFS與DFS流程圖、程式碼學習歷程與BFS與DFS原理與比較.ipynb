```python
# Python3 Program to print BFS traversal 
# from a given source vertex. BFS(int s) 
# traverses vertices reachable from s. 
from collections import defaultdict        
```

```python
s = [('red', 1), ('blue', 2), ('red', 3), ('blue', 4), ('red', 1), ('blue', 4)]
d = defaultdict(list)

for k, v in s:
    d[k].append(v)
    
print(d)
```

```python
s = [('red', 1), ('blue', 2), ('red', 3), ('blue', 4), ('red', 1), ('blue', 4)]
d = defaultdict(set)

for k, v in s:
    d[k].add(v)
    
print(d)
```

```python
s = 'mississippi'
d = defaultdict(int)
for k in s:
    d[k] += 1
print(d)
```

### CODE1 _éæ­¸æ–¹å¼

```python
# This class represents a directed graph 
# using adjacency list representation 
class Graph:
    # Constructor 
    def __init__(self): 
        # default dictionary to store graph 
        self.graph = defaultdict(list) 

    # function to add an edge to graph 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
  
    # Function to print a BFS of graph 
    def BFS(self, s): 
        """
        :type s: int
        :rtype: list
        """
        queue = []
        color = [False]*(len(self.graph)) #è¨­[False]*self.graphçš„é•·åº¦
        
        queue.append(s)  #å…ˆå°‡èµ·é»sæ”¾é€²queue[]
        color[s] = True  #colorè£¡çš„æ•¸ä»£è¡¨é‚£å€‹å€¼å·²ç¶“èµ°è¨ªéï¼Œèµ°è¨ªéçš„å€¼è®Šç‚ºTrue
        
        while queue:         #ç•¶queue == Trueï¼Œmove on
            s = queue.pop(0) #å°‡queue->`stack`è£¡æœ€å¾Œæ”¾é€²çš„å€¼popå‡º
            print(s)
            
            for i in self.graph[s]:    #è¨­ä¸€å€‹forè¿´åœˆ
                if color[i] == False:  #å¦‚æœcolorè£¡çš„ï½‰ä½ç½®ç‚ºè¢«èµ°è¨ªéï¼Œmove on
                    queue.append(i)    #å°‡iæ”¾é€²queue[] 
                    color[i] = True    #æ‰€ä»¥colorè£¡çš„iä½ç½®å°±æœƒè®Šç‚ºTrue
                
    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """
        color = [False]*(len(self.graph))  #è¨­[False]*self.graphçš„é•·åº¦
        
        #è¨­ä¸€å€‹æ–°çš„defï¼Œä»¥éæ­¸çš„æ–¹å¼å‘¼å«
        def dfs(s):
            color[s] = True  #æ”¾å…¥dfs(s)è£¡çš„sä½ç½®æœƒè®ŠæˆTrue      
            print(s)
            for j in self.graph[s]:    #è¨­ä¸€å€‹forè¿´åœˆ
                if color[j] == False:  #å¦‚æœcolorè£¡jçš„ä½ç½®çš„å€¼æ˜¯False
                    dfs(j)             #å‘¼å«dfs(j)ï¼Œé€™æ¨£å°±æœƒæŠŠcolorä¸­çš„jçš„ä½ç½®è®ŠæˆTrue
        dfs(s) #å‘¼å«dfs(s)
        return
```

```python
g = Graph() 
g.addEdge(0, 1) 
g.addEdge(0, 2) 
g.addEdge(1, 2) 
g.addEdge(2, 0) 
g.addEdge(2, 3) 
g.addEdge(3, 3) 

print(g.BFS(2))
print(g.DFS(2))
```

```python
from collections import defaultdict     
class Graph:

    def __init__(self): 
        
        self.graph = defaultdict(list) 
    
    def addEdge(self,u,v): 
        
        self.graph[u].append(v) 
            
    def BFS(self, s): 
        """
        :type s: int
        :rtype: list
        """
        color = []      
        #color = set()  #è¨­colorç‚ºä¸€å€‹set()
        queue = [s]     #queueç‚ºä¸€å€‹å·²æœ‰èµ·é»sçš„list
        #res = [s]      #è¨­ä¸€å€‹å·²æœ‰èµ·é»sçš„listï¼Œå¥½æŠŠå·²å®Œæˆèµ°è¨ªçš„å€¼æ”¾å…¥
        print("BFS")
        while queue:
            top = queue.pop(0) #å°‡queueè£¡çš„å€¼popå‡ºç‚ºtop
            
            if top not in color:
                color.append(top) 
                #color.add(top)
                #colorå› ç‚ºæ˜¯set()ï¼Œä¸æ˜¯list()ï¼Œæ‰€ä»¥ä½¿ç”¨addå°‡topåŠ å…¥
                queue.extend(set(self.graph[top]) - set(color))
                #å°‡self.graph[top]ä»¥set()çš„æ–¹å¼å‘ˆç¾ï¼Œåœ¨æ¸›æ‰colorå¾Œï¼ˆå…©è€…çš„å·®é›†ï¼‰ï¼Œå°‡ä¸¦æœªèµ°è¨ªéçš„å€¼æ”¾é€²queueè£¡
                #res.extend(set(self.graph[top]) - color)
                #å› ç‚ºsetæœƒè®“æ•¸å­—è‡ªå‹•ç”±å°åˆ°å¤§æ’åˆ—ï¼Œæ‰€ä»¥ç„¡æ³•return colorå–å¾—bfsï¼Œæ‰€ä»¥è¨­ä¸€å€‹res[s]
                #å°‡å³å°‡è¦èµ°è¨ªçš„å€¼ä»¥extendçš„æ–¹å¼æ”¾é€²res
        return color

    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """
        color = list() #è¨­colorç‚ºä¸€å€‹list()
        queue = [s]    #queueç‚ºä¸€å€‹å·²æœ‰èµ·é»sçš„list
        print("DFS")
        #è¨­ä¸€å€‹æ–°çš„defï¼Œä»¥éæ­¸çš„æ–¹å¼å‘¼å«
        def dfs(s):
            
            if queue:  #å¦‚æœqueue == Trueï¼Œmove on
                top = queue.pop() #å°‡queueçš„æœ€ä¸Šå±¤çš„å€¼popç§»å‡º

                if top not in color:   #è‹¥æ˜¯topé€™å€‹å€¼ä¸¦æœªè¢«æ”¾å…¥colorè£¡ï¼Œmove on
                    color.append(top)  #top appendé€²color
                    i = queue.extend(set(self.graph[top]) - set(color)) 
                    #å°‡self.graph[top]ä»¥set()çš„æ–¹å¼å‘ˆç¾ï¼Œåœ¨æ¸›æ‰set(color)å¾Œï¼Œå°‡ä¸¦æœªèµ°è¨ªéçš„å€¼æ”¾é€²queueè£¡
                    #print(set(self.graph[top]) - set(color))
                    dfs(i) #å‘¼å«dfs(i)ï¼Œæ­¤æ­¥é©Ÿæ˜¯ç‚ºäº†æŒçºŒåŸ·è¡Œï¼Œwhileè¿´åœˆ
                    
        dfs(s)  #å‘¼å«dfs(s)
        return color
```

### BFS_Code4

```python
def BFS(self, s): 

    color = set()   #è¨­colorç‚ºä¸€å€‹set()
    queue = [s]     #queueç‚ºä¸€å€‹å·²æœ‰èµ·é»sçš„list
    res = [] 
    def bfs(s):
        if queue:
            top = queue.pop(0) #pop index0
            res.append(top)    #å°‡queueè£¡çš„æœ€åº•å±¤æ•¸ï¼Œappendè‡³æ–°çš„res[]

            if top not in color:
                color.add(top) 
                k = queue.extend(set(self.graph[top]) - color)
                bfs(k)
    bfs(s)
    return res
```
