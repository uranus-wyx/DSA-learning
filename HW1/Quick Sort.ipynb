> Step 4ï¼šå®šç¾©Quicksortå‡½æ•¸ï¼Œå¦‚æœå·¦é‚Š<å³é‚Šï¼Œæ‰¾å‡ºpivoté€²è¡Œåˆ†å€ï¼Œä»¥äºŒåˆ†æ³•åˆ†æ”¯ä¸‹å»ï¼Œæœ€çµ‚é”åˆ°æ²’æœ‰å…ƒç´ æ˜¯left>right

```python
def partition(array, left, right):
    i = left-1
    for j in range (left, right):
        if array[j] <= array[right]:
            i +=1
            array[j] , array[i] = array[i] ,  array[j]
    array[i+1],array[right] = array[right] , array[i+1]
    return i+1 

def quicksort (array, left, right):
    
    if left < right:
        pivot = partition(array, left, right)
        quicksort (array, left, pivot-1)
        quicksort (array, pivot+1, right)
```

```python
array = [5,6,7,2,3,4,1,0]
quicksort(array, 0, 7)
print(array)
```

### ä¸‹é¢æ˜¯ä»¥.append()çš„æ–¹å¼å»å°‡listè£¡çš„å…ƒç´ èˆ‡pivotæ¯”è¼ƒï¼Œé€²è€Œæ”¾å…¥ larger[] & smaller[] & mid[] çš„listè£¡é¢ï¼Œæœ€å¾Œå†ä»¥Quicksort()å‡½å¼ï¼ŒæŒçºŒå‘¼å«è‡ªå·±çš„å‡½å¼ï¼Œå†å°‡ä¸‰è€…ç›¸åŠ ã€‚

```python
def Quicksort(list):
    smaller = []
    larger = []
    mid = []
    
    if len(list)<=1:
        return list
    else:
        pivot = list[0]
        for i in list:
            if i < pivot:
                smaller.append(i)
            elif i > pivot:
                larger.append(i)
            else:
                mid.append(pivot)
            
    larger = Quicksort(larger)
    smaller = Quicksort(smaller)
    return (smaller + mid + larger)
        
```

```python
A = [5,6,7,2,3,4,1,0]
Quicksort(A)
```
