```python
class Node:

    def __init__(self, val, nextNode=None):
        """
        initialize your data structure here.
        """
        self.val = val #valæ˜¯ä¸€å€‹å€¼
        self.temp_min = val #æš«æ™‚æœ€å°å€¼
        self.next = nextNode #ä¸‹ä¸€å€‹å€¼
```

```python
class MinStack:
    
    def __init__(self):
        
        self.topNode = None #å®¹å™¨ä¸­æ²’æœ‰æ•¸å€¼çš„å­˜åœ¨

    def push(self, x: int) -> None:
        
        if self.topNode is None:
            self.topNode = Node(x, None) #å¦‚æœæ²’æœ‰ç¬¬ä¸€å€‹å€¼çš„è©±ï¼ŒåŠ é€²ä¸€å€‹Node
        else:
            temp = self.topNode.temp_min  #æœ€ä¸Šé¢çš„æœ€å°å€¼æˆç‚ºtemp
            self.topNode = Node(x, self.topNode) #åŠ é€²ä¸€å€‹Nodeæˆç‚ºæœ€ä¸Šé¢çš„æ•¸å€¼
            if temp < self.topNode.val:
                self.topNode.temp_min = temp    #å¦‚æœåŠ é€²ä¾†çš„æ•¸å€¼æ²’æœ‰æ¯”åŸä¾†çš„æœ€å°å€¼å°ï¼Œé‚£éº¼æœ€å°å€¼ä¾èˆŠæ˜¯temp

            
    def pop(self) -> None:
        
        self.topNode = self.topNode.next  #popæ‰æœ€ä¸Šé¢çš„é‚£ä¸€å€‹æ•¸å­—ï¼Œé‚£éº¼åŸæœ¬topnodeçš„ä¸‹ä¸€å€‹ç¯€é»å°‡æˆç‚ºæ–°çš„é ‚é»


    def top(self) -> int:
        
        return self.topNode.val  #å›å‚³é ‚é»çš„å€¼
        

    def getMin(self) -> int:
        
        return self.topNode.temp_min
```

```python
obj = MinStack()
obj.push(-2);
obj.push(0);
obj.push(-3);
obj.pop()
param_3 = obj.top()
param_4 = obj.getMin()
```
