```python
def dijkstra(graph,src):
    # åˆ¤æ–­å›¾æ˜¯å¦ä¸ºç©ºï¼Œå¦‚æœä¸ºç©ºç›´æ¥é€€å‡º
    if graph is None:
        return None
    nodes = [i for i in range(len(graph))]  # è·å–å›¾ä¸­æ‰€æœ‰èŠ‚ç‚¹
    visited=[]  # è¡¨ç¤ºå·²ç»è·¯ç”±åˆ°æœ€çŸ­è·¯å¾„çš„èŠ‚ç‚¹é›†åˆ
    if src in nodes:
        visited.append(src)
        nodes.remove(src)
    else:
        return None
    distance={src:0}  # è®°å½•æºèŠ‚ç‚¹åˆ°å„ä¸ªèŠ‚ç‚¹çš„è·ç¦»
    for i in nodes:
        distance[i]=graph[src][i]  # åˆå§‹åŒ–
    # print(distance)
    path={src:{src:[]}}  # è®°å½•æºèŠ‚ç‚¹åˆ°æ¯ä¸ªèŠ‚ç‚¹çš„è·¯å¾„
    k=pre=src
    while nodes:
        mid_distance=float('inf')
        for v in visited:
            for d in nodes:
                new_distance = graph[src][v]+graph[v][d]
                if new_distance < mid_distance:
                    mid_distance=new_distance
                    graph[src][d]=new_distance  # è¿›è¡Œè·ç¦»æ›´æ–°
                    k=d
                    pre=v
        distance[k]=mid_distance  # æœ€çŸ­è·¯å¾„
        path[src][k]=[i for i in path[src][pre]]
        path[src][k].append(k)
        # æ›´æ–°ä¸¤ä¸ªèŠ‚ç‚¹é›†åˆ
        visited.append(k)
        nodes.remove(k)
        print(visited,nodes)  # è¾“å‡ºèŠ‚ç‚¹çš„æ·»åŠ è¿‡ç¨‹
    return distance,path
if __name__ == '__main__':
    graph_list = [ [0, 2, 1, 4, 5, 1],
            [1, 0, 4, 2, 3, 4],
            [2, 1, 0, 1, 2, 4],
            [3, 5, 2, 0, 3, 3],
            [2, 4, 3, 4, 0, 1],
            [3, 4, 7, 3, 1, 0]]

    distance,path= dijkstra(graph_list, 0)  # æŸ¥æ‰¾ä»æºç‚¹0å¼€å§‹å¸¦å…¶ä»–èŠ‚ç‚¹çš„æœ€çŸ­è·¯å¾„
    print(distance)
```

```python
def startwith(start: int, mgraph: list) -> list:
    passed = [start]
    nopass = [x for x in range(len(mgraph)) if x != start]
    dis = mgraph[start]
    
    while len(nopass):
        idx = nopass[0]
        for i in nopass:
            if dis[i] < dis[idx]: idx = i

        nopass.remove(idx)
        passed.append(idx)

        for i in nopass:
            if dis[idx] + mgraph[idx][i] < dis[i]: 
                dis[i] = dis[idx] + mgraph[idx][i]
    return dis


if __name__ == "__main__":
    inf = 10086
    mgraph = [[0, 1, 12, inf, inf, inf],
              [inf, 0, 9, 3, inf, inf],
              [inf, inf, 0, inf, 5, inf],
              [inf, inf, 4, 0, 13, 15],
              [inf, inf, inf ,inf, 0, 4],
              [inf, inf, inf, inf ,inf, 0]]

    dis = startwith(0, mgraph)
    print(dis)
```
