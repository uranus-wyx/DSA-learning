```python
nums = [1,2,3,3,5]
```

```python
class Solution:
    def findErrorNums(self,nums):
        """
        :type nums:List[int]
        :rtype:List[int]
        """
        holder=[0]*len(nums)
        repeated_num = 0
        
        for i in range(0, len(nums)):
            if holder[nums[i]-1]!=0:
                repeated_num=nums[i]
            holder[nums[i]-1]=nums[i]
        return [repeated_num, holder.index(0)+1]
```

```python
a=Solution()
```

```python
print(a.findErrorNums(nums))
```

```python
def findErrorNums(nums):

        N = len(nums)
        count = [0]* (N+1)
        for x in nums:
            count[x] += 1        
        for x in range(1,N+1):
            if count[x] == 2:
                dup = x
            if count[x] == 0:
                missing = x
                
        return [dup,missing]
```

```python
print (findErrorNums(nums))
```
